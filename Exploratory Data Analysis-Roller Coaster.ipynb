```python
df = df[['coaster_name', # 'Length', 'Speed', 
    'Location', 'Status', #'Opening date',
       #'Type', 
    'Manufacturer',
    #'Height restriction', 'Model', 'Height',
      # 'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
      # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
      # 'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 
    'Opened', 
    #'Replaced by', 'Website',
     #  'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
      # 'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced',
    'latitude', 'longitude', 
    'Type_Main',
       'opening_date_clean', 
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
   # 'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
# rename column

df = df.rename(columns = {'coaster_name': 'Coaster_Name',
                    'year_introduced' : 'Year_introduced',
                    'latitude' : 'Latitude',
                    'longitude' : 'Longitude',
                    'opening_date_clean' : 'Opening_date',
                    'speed_mph' : 'Speed_mph',
                    'height_ft' : 'Height_ft',
                     'Inversions_clean' : 'Inversions',
                    'Gforce_clean' : 'Gforce'
                    })
```

```python
df = df.loc[~df.duplicated(subset = ['Coaster_Name' , 'Location', 'Opening_date'])].reset_index(drop = True).copy()
```

```python
# implementing Univariate analysis on Year_introduced numerical column

df['Year_introduced'].value_counts()
```

```python
ax = df['Year_introduced'].value_counts().head(10).plot(kind = 'bar', title = 'Top 10 years Coasters introduced')

ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
```

```python
ax1 = df['Speed_mph'].plot(kind = 'hist', bins = 25, title = 'Coaster speed (mph)')
ax1.set_xlabel('Speed (mph)')
```

```python
ax2 = df['Speed_mph'].plot(kind = 'kde', title = 'Coaster speed (mph)')
ax2.set_xlabel('Speed (mph)')
```

```python
df.plot(kind ='scatter' , x = 'Speed_mph' , y ='Height_ft', title = 'Coaster Speed vs. Height')
plt.show()
```

```python
sns.scatterplot(x = 'Speed_mph' , 
                y ='Height_ft',
                hue = 'Year_introduced',
                data = df)
```

```python
# Comparing different features and their relationships
sns.pairplot(df, vars = ['Year_introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce'],
            hue = 'Type_Main')
plt.show()
```

```python
df_corr = df[['Year_introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce']].dropna().corr()
```

```python
sns.heatmap(df_corr, annot = True)
```

```python
# What are the locations with the fastest roller coasters(minimum od 10)?

df['Location'].value_counts()
```

```python
df.query('Location !="Other"').groupby('Location')['Speed_mph'].agg(['mean', 'count']).query('count >= 10').sort_values('mean')['mean'].plot(kind = 'barh', figsize = (12,5), title = ' Average coaster speed by location')

plt.show()
```
